```python
import pandas as pd
import numpy as np

df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(1)
```

# Genera una tabla dinámica que muestre el peso promedio de las personas por estado.

```python
pd.pivot_table(df,
               index = 'desc_ent',
               values = 'peso',
              aggfunc = 'mean')
```

# Agrega columnas que muestran el peso promedio por sexo por estado.

```python
df.rename(columns={'desc_ent':'Estado',
                  },
         inplace = True)

pd.pivot_table(df,
               index = ['Estado','sexo'],
               values = 'peso',
              aggfunc = 'mean')
```

# La variable 'sexo' tiene valores discretos 1 y 2. Usando la información del peso promedio, sugiere qué número corresponde a hombre y cuál a mujer.

Viendo los resultados de la tabla anterior podemos suponer que en la variable 'sexo' el numero 1 representa al hombre y el 2 a la mujer, ya que se puede suponer que el peso menor es el de la mujer.
