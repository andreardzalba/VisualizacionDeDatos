```python
import matplotlib.pyplot as plt
import pandas as pd

dataframe = pd.read_csv("Police_Department_Incident_Reports__2018_to_Present.csv")
dataframe['count']=dataframe.groupby('Incident Day of Week')['Incident Day of Week'].transform('count')
dataframe.drop_duplicates('Incident Day of Week',inplace=True)
plt.figure(figsize=(50,8))
ax1 = plt.subplot(121,aspect='equal')
dataframe.plot(kind = 'pie', y = 'count',ax=ax1,autopct='%1.1f%%',startangle=90,shadow=False,labels=dataframe['Incident Day of Week'],legend=False,fontsize=10)
print("Porcentaje de incidentes por día de la semana")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

dataframe = pd.read_csv("Police_Department_Incident_Reports__2018_to_Present.csv")
dataframe['count']=dataframe.groupby('Incident Year')['Incident Year'].transform('count')
dataframe.drop_duplicates('Incident Year',inplace=True)
plt.figure(figsize=(50,8))
ax1 = plt.subplot(121,aspect='equal')
dataframe.plot(kind = 'pie', y = 'count',ax=ax1,autopct='%1.1f%%',startangle=90,shadow=False,labels=dataframe['Incident Year'],legend=False,fontsize=10)
print("Porcentaje de incidentes por año")
plt.show()
```
