```python
x = np.array([
    [1,2],
    [3,4]
],
dtype = np.float64)

y = np.array([
    [5,6],
    [7,8]
],
dtype = np.float64)

print('El matriz x es: \n',x)
print('El matriz y es: \n',y)
```

```python
print(np.array_equal(x+y, np.add(x,y)))
```

```python
print(x-y)
```

```python
print(np.array_equal(x-y,np.subtract(x,y)))
```

```python
print(np.array_equal(x*y,np.multiply(x,y)))
```
