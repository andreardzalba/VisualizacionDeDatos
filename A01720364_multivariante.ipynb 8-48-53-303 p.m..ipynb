## Generar histogramas de las variables de presión sanguínea 'sistol' y 'diastol' y haz que se muestren de manera contigua

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('datasets/antropometria-dataset.csv')

df.head(3)
```

```python
df.dtypes
```

```python
plt.subplot(1,2,1)
df['sistol'].plot(kind='hist')
plt.title('Sistol')

plt.subplot(1,2,2)
df['diastol'].plot(kind='hist')
plt.title('Diastol')

plt.show()
```

## Escribe el código en Python para generar histogramas de las variables 'peso' y 'cintura' y haz que se muestren de manera contigua.

```python
plt.subplot(1,2,1)
df['peso'].plot(kind='hist')
plt.title('Peso')

plt.subplot(1,2,2)
df['cintura'].plot(kind='hist')
plt.title('Cintura')

plt.show()
```

## Escribe el código en Python para generar un diagrama de dispersión para analizar la relación entre las variables 'peso' y 'cintura'.

```python
sns.lmplot(
    x = 'peso',
    y = 'cintura',
    data = df,
)

plt.title("Diagrama de dispersión de peso vs. cintura ")

plt.show()
```

```python
hombres = df.loc[df['sexo']==1] 
mujeres = df.loc[df['sexo']==2]
pesohombres = hombres['peso']
cinturahombres = hombres['cintura']
pesomujeres = mujeres['peso']
cinturamujeres = mujeres['cintura']
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = pesohombres, y = cinturahombres)
plt.xlabel("'peso'")
plt.ylabel("'cintura'")
plt.title("Diagrama de dispersión de peso vs. cintura en hombres")

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = pesomujeres, y = cinturamujeres)
plt.xlabel("'peso'")
plt.ylabel("'cintura'")
plt.title("Diagrama de dispersión de peso vs. cintura en mujeres")

plt.show()
```
