```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')

df = pd.read_excel('datasets/Canada.xlsx',
                  sheet_name = 'Canada by Citizenship',
                  skiprows = range(20),
                  skipfooter = 2,
                  engine ='openpyxl')

df.head(2)
```

```python
df.shape
```

```python
years =  list(map(str,range(1980,2014)))
#years
```

```python
df.columns = list(map(str,df.columns))
```

```python
df_t = pd.DataFrame(df[years].sum(axis=(0)))
df_t.head(2)
```

```python
df_t.index = map(float,df_t.index)
df_t.reset_index(inplace = True)
df_t.head(2)
```

```python
df_t.columns = ['year','total']
df_t.tail(2)
```

```python
#puedes no poner lo de reg_plot nomas es apra guardarlo en una variable
plt.figure(figsize=(15,10))
sns.set(font_scale=1.5)
sns.set_style('whitegrid') #ticks, mas limpio

reg_plot = sns.regplot(x = 'year',
                      y ='total',
                      data = df_t,
                      color ='b',
                      marker = '+',
                      scatter_kws = {'s':150})

reg_plot.set(xlabel = 'year',
            ylabel = 'total imigration')

reg_plot.set_title('Total imigration to Canada from 1980-2013')
plt.show()
```

```python
df = pd.read_csv('datasets/headbrain.csv')
df.head(2)
```

```python
a = df['Head Size(cm^3)']
b = df['Brain Weight(grams)']

#print(a)
#print(b)
```

```python
np.corrcoef(a,b)
```

```python
plt.scatter(a,b,
           c = 'y',
           label = 'Data points')

plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm

model = sm.OLS.from_formula('a~b', #b = b0 +b1*a
                           data=df)
result = model.fit()
result.summary()
```

```python
from sklearn import linear_model

lm = linear_model.LinearRegression()
X = pd.DataFrame(df['Head Size(cm^3)'])
Y = pd.DataFrame(df['Brain Weight(grams)'])

model_lm = lm.fit(X,Y)
model_lm
```

```python
model_lm.intercept_
```

```python
model_lm.coef_
```

```python
model_lm.score(X,Y)
```

```python
brain_weight = pd.DataFrame([200])
predict_headsize = model_lm.predict(brain_weight)
predict_headsize
```

```python
df.plot(kind = 'scatter',
       x = 'Head Size(cm^3)',
       y = 'Brain Weight(grams)',
       color = 'r')

plt.plot(X,
        model_lm.predict(X),
        linewidth = 2,
        color = 'g')

plt.show()
```
